```python
#Creating data points

def data_creation(n=100):
    X=tf.random.uniform(shape=(n,))
    print(X)
    return X
#plot variable that contains the random tesnors
plot=data_creation(n=1000)
```

```python
print("Now we will add noise to our data with bias and weight as well\n 100 data only")
def data_creation(w=0.1,b=0.2,n=100):
    X=tf.random.uniform(shape=(n,))
    noise=tf.random.normal(shape=(n,),stddev=0.1 )
    #normal distribution seconf parameter is standard deviation
    y=X*w+b+noise
    print(y)
    return X,y

X,Y=data_creation(n=100)
```

```python
print("Again plotting with noise")

plt.hist(Y.numpy())
plt.show()
```

```python
print("Plotting the data\n") 

plt.plot(X.numpy(),Y.numpy(),"b.")
```
