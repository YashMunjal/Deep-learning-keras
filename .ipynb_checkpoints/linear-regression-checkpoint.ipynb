```python
print("Now we will add noise to our data with bias and weight as well\n 200 data only")
def data_creation(w=0.1,b=0.2,n=100):
    X=tf.random.uniform(shape=(n,))
    noise=tf.random.normal(shape=(n,),stddev=0.05 )
    #normal distribution seconf parameter is standard deviation
    y=X*w+b+noise
    print(y)
    return X,y

X,Y=data_creation(n=200)
```

```python
print("Now in this data lets make a new data line with the an ideal solution line")
w=0.1
b=0.2
plt.plot(X.numpy(),Y.numpy(),"b.")
plt.plot([0,1],[0*w+b,1*w+b],"g:")
```

```python
print("Now we lets say we need to find the error between the data and the optimal soultion the green line\n")
print("So here we will use the solution line in red to find the solution")

w_guess=0.0
b_guess=0.0
plt.plot(X.numpy(),Y.numpy(),"b.")
plt.plot([0,1],[0*w+b,1*w+b],"g:")
plt.plot([0,1],[0*w_guess+b_guess,1*w_guess+b_guess],"r:")
```

```python
print("So we form a prediction function\n")

def predict(x,w=w,b=b):
    y=w*x+b
    return y
```

```python
#Now a loss function to tell the difference between the distance between points  and prediction like

#Helper function
#simply tells how far the error is
def mean_squared_error(y_pred,y):
    return tf.reduce_mean(tf.square(y_pred-y))
```

```python
print(mean_squared_error(predict(X),Y))
```

```python
print("But random guessing can take forever. So\n")

w=tf.Variable(0.0)
b=tf.Variable(0.0)
history = []
learning_rate = 0.7 
#should be kept minimum for less delta b
#we need derivatives d/dx
#I kept it big as my prediction was coming 100% true
steps=200

for step in range(steps):
    with tf.GradientTape() as tape:
        loss=mean_squared_error(predict(X,w=w,b=b),Y)
        
    gradients = tape.gradient(loss, [w, b])
    
    history.append((w.numpy(), b.numpy(), loss.numpy()))
    w.assign_sub(gradients[0] * learning_rate)
    b.assign_sub(gradients[1] * learning_rate)
    
    if step % 20 == 0:
        print("Step %d, Loss %f" % (step, loss.numpy()))
```
